## calculate state wise:
- A:Total number of population'
- B:Total number of population with different religions.

```python
data.groupby('State_name').Population.sum()
```

```python
data.groupby('State_name').Population.sum().sort_values()
```

```python
import matplotlib.pyplot as plt
# Bar chart with day against tip
bar=plt.bar(data['State_name'], data['Population'])
 
plt.title("State wise Population")
 
# Setting the X and Y labels
plt.xlabel('State')
plt.ylabel('Population')
 
# Adding the legends
plt.show(bar)
```
